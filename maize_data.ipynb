```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("MaizeDataset.xlsx")
df.columns
```

```python
df_county = df.groupby('County')
```

```python
df = pd.DataFrame(columns=['County', 'Region', 'State/Province', '14C age', 'Latitude', 'Longitude', 'Location reference', 'Sources'])

for name, group in df_county:

    group['County'] = name
    group.sort_values('14C age', inplace=True)
    if group.Latitude.notna().any():
        group['Latitude'] = group.dropna(subset=["Latitude"])["Latitude"].values[0]
    if group.Longitude.notna().any():
        group['Longitude'] = group.dropna(subset=["Longitude"])["Longitude"].values[0]
    if group['Location reference'].notna().any():
        group['Location reference'] = group.dropna(subset=["Location reference"])["Location reference"].values[0]
    df = pd.concat([df, group.iloc[[0]][['County', 'Region', 'State/Province', '14C age', 'Latitude', 'Longitude', 'Location reference','Sources']]])

df["Year"] = 1950 - df["14C age"]
```
